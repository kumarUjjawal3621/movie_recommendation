```python
import re
import pandas as pd
movie_df=pd.DataFrame()

for e in movie_data:
    try:
        movie_name='NA'
        year='NA'
        duration='NA'
        rating='NA'
        vote_cnt='NA'
        summary='NA'
        
        original_string=e.h3.text
        split_parts = original_string.split('. ')
        movie_name = split_parts[1] #if len(split_parts) > 1 else original_string
        
        year_duration=e.find_all('span', class_='sc-1e00898e-8 hsHAHC dli-title-metadata-item')
        year=year_duration[0].text
        if len(year_duration)>1:
            duration=year_duration[1].text
        
        input_string=e.find('span', class_='ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating').text
        pattern = r'([^\xa0]+)\xa0\(([^\)]+)\)'
        match = re.search(pattern, input_string)
        if match:
            rating = match.group(1)
            vote_cnt = match.group(2)
        para=e.find('div', class_='sc-73c670dc-2 djRAjt').text
        index_of_votes = para.rfind('Votes')
        
        if index_of_votes != -1:
            summary = para[:index_of_votes]
        else:
            print("The paragraph does not contain 'Votes'.")

        temp_df={'Name':[movie_name],'Year':[year],'Duration':[duration],'Rating':[rating],'Votes':[vote_cnt],'Description':[summary]}
        temp_df=pd.DataFrame(temp_df)
        movie_df=pd.concat([movie_df,temp_df])
    except :
        print('failed')
    
movie_df
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
# Locate all the buttons with the specified class name
buttons = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.CLASS_NAME, 'sc-73c670dc-5'))
)
print(len(buttons))
```

```python
import time
df=pd.DataFrame()

# Click each button one by one
i=0
while i<len(buttons):
    b=0
    try:
        
        button=buttons[i]
        time.sleep(.5)
        button.click()
#         time.sleep(.1)
        b=1
        # Wait for the main content to be present
        main_content = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'sc-7316798c-0.koKMMU')))

        # Extract genre
        genres_elements = main_content.find_elements(By.CSS_SELECTOR, 'ul[data-testid="btp_gl"] li')
        genres_list = [genre.text.strip() for genre in genres_elements]

        # Extract director
        director_element = main_content.find_element(By.CSS_SELECTOR, 'div[data-testid="p_ct_dr"] a.ipc-link.ipc-link--baseAlt.sc-9bca7e5d-0.ktChvQ')
        director_name = director_element.text.strip() if director_element else None

        # Extract cast
        cast_elements = main_content.find_elements(By.CSS_SELECTOR, 'div[data-testid="p_ct_cst"] a.ipc-link.ipc-link--baseAlt.sc-9bca7e5d-0.ktChvQ')
        cast_names = [cast.text.strip() for cast in cast_elements]

        print(genres_list,director_name,cast_names)

        temp = {'Director': [director_name], 'Cast': [cast_names], 'Genre': [genres_list]}
        print(temp)
        df = pd.concat([df, pd.DataFrame(temp)], ignore_index=True)  # Assign the result back to df
 
        
        b = driver.find_element(By.CLASS_NAME,'ipc-promptable-base__close')
        b.click()
        b=0

        print('complete')
        i+=1
            
    except Exception as e:
        print(i,e)
        if b==1:
            b = driver.find_element(By.CLASS_NAME,'ipc-promptable-base__close')
            b.click()
```

```python
import time
df=pd.DataFrame()

# Click each button one by one
i=0
while i<len(buttons):
    b=0
    try:
        
        button=buttons[i]
#         time.sleep(10)
        button.click()
#         time.sleep(2)
        b=1

        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')

        # Find the director's name
        director_data = soup.find('div', {'data-testid': 'p_ct_dr'}).find('a', {'class': 'ipc-link ipc-link--baseAlt sc-9bca7e5d-0 ktChvQ'})
        director_name = director_data.text.strip() if director_data else None

        # Find the stars' names
        stars_data = soup.find('div', {'data-testid': 'p_ct_cst'}).find_all('a', {'class': 'ipc-link ipc-link--baseAlt sc-9bca7e5d-0 ktChvQ'})
        stars_names = [star.text.strip() for star in stars_data] if stars_data else []

        genres_data = soup.find('ul', {'data-testid': 'btp_gl'})
        genres_list = [genre.text.strip() for genre in genres_data.find_all('li')] if genres_data else []

        temp = {'Director': [director_name], 'Cast': [stars_names], 'Genre': [genres_list]}
        print(temp)
        df = pd.concat([df, pd.DataFrame(temp)], ignore_index=True)  # Assign the result back to df
 
        
        b = driver.find_element(By.CLASS_NAME,'ipc-promptable-base__close')
        b.click()
        b=0

        print('complete')
        i+=1
            
    except Exception as e:
        print(i,e)
        if b==1:
            b = driver.find_element(By.CLASS_NAME,'ipc-promptable-base__close')
            b.click()
```
